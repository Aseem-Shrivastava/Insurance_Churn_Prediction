## Modelling

```python
# Importing relevant packages

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score, precision_score, recall_score, f1_score, make_scorer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
```

```python
X = df1.drop(columns=["labels"])
y = df1["labels"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21, stratify=y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Checking different classification models with base parameters

models = {"Logistic Regression": LogisticRegression(),
          "SVC": SVC(),
          "NAive Bayes": GaussianNB(),
          "Decision Tree": DecisionTreeClassifier(),
          "Gradient Boost": GradientBoostingClassifier(),
          "XGB": XGBClassifier()}

metrics = {"Accuracy": make_scorer(accuracy_score),
           "Precision": make_scorer(precision_score, average="weighted"),
           "Recall": make_scorer(recall_score, average="weighted"),
           "F1": make_scorer(f1_score, average="weighted")}

for name, model in models.items():
    print(f"-----{name}-----")
    model.fit(X_train, y_train)
    
    pred = model.predict(X_test)
    for metric_name, metric in metrics.items():
        print(f"Test {metric_name}: {round(metric._score_func(y_test, pred),2)}")
    
    cm = confusion_matrix(y_test, pred)
    cm_disp = ConfusionMatrixDisplay(cm, display_labels=["Churn", "No Churn"])
    cm_disp.plot()
    plt.show()
```

```python
# Checking cross-validation scores for different classification models with base parameters
from sklearn.model_selection import cross_val_score

results={}
for model_name, model in models.items():
    print(f"-----{model_name}-----")
    results[model_name] = {}
    for metric_name, metric in metrics.items():
        scores = cross_val_score(model, X_train, y_train, cv=3, scoring=metric, n_jobs=-1)
        results[model_name][metric_name] = scores.mean()
        print(f"{metric_name}: {round(scores.mean(),2)}")
    print()
```

```python
# Selecting XGBoost Classifier for hyper parameter tuning as it performs best in the cross_val_score across different metrics

from sklearn.model_selection import GridSearchCV

metrics = {
    "Accuracy": make_scorer(accuracy_score),
    "Precision": make_scorer(precision_score, average="weighted", zero_division=1),
    "Recall": make_scorer(recall_score, average="weighted", zero_division=1),
    "F1": make_scorer(f1_score, average="weighted", zero_division=1)
}

params = {"n_estimators": [100, 300, 500],
          "max_depth": [1,3,6,9],
          "learning_rate": [0.01, 0.1, 0.2],
          "subsample": [0.8, 1.0]}


grid_search = GridSearchCV(estimator=XGBClassifier(), param_grid=params, scoring=metrics, refit="F1", cv=3, n_jobs=-1)

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print("Best Parameters:", best_params)
```

```python
# Using the tuned model for prediction

tuned_xgb = grid_search.best_estimator_

pred = tuned_xgb.predict(X_test)

for metric_name, metric in metrics.items():
    print(f"{metric_name}: {round(metric._score_func(y_test, pred),2)}")

cm = confusion_matrix(y_test, pred)
cm_disp = ConfusionMatrixDisplay(cm, display_labels=["Churn", "No Churn"])
cm_disp.plot()
```

```python
# Tuned model performs poor vs the base XGBClassifier. This could be due to class imbalance.

# Oversampling using SMOTE
from imblearn.over_sampling import SMOTE

oversample = SMOTE()
X, y = oversample.fit_resample(X,y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21, stratify=y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Hyper parameter tuning the XGBClassifier model on the over sampled data and then fitting the best estimator

metrics = {
    "Accuracy": make_scorer(accuracy_score),
    "Precision": make_scorer(precision_score, average="weighted", zero_division=1),
    "Recall": make_scorer(recall_score, average="weighted", zero_division=1),
    "F1": make_scorer(f1_score, average="weighted", zero_division=1)
}

params = {"n_estimators": [100, 300, 500],
          "max_depth": [1,3,6,9],
          "learning_rate": [0.01, 0.1, 0.2],
          "subsample": [0.8, 1.0]}


grid_search = GridSearchCV(estimator=XGBClassifier(), param_grid=params, scoring=metrics, refit="F1", cv=3, n_jobs=-1)

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print("Best Parameters:", best_params)


tuned_xgb = grid_search.best_estimator_

pred = tuned_xgb.predict(X_test)

for metric_name, metric in metrics.items():
    print(f"{metric_name}: {round(metric._score_func(y_test, pred),2)}")

cm = confusion_matrix(y_test, pred)
cm_disp = ConfusionMatrixDisplay(cm, display_labels=["Churn", "No Churn"])
cm_disp.plot()
```
